```python
import json

with open('../data/all_sentences.json', encoding='utf-8') as f:
    all_sentences = json.load(f)

with open('../data/all_sentences_translation.json', encoding='utf-8') as f:
    all_sentences_translation = json.load(f) 

with open('../data/gpt_generate.json', encoding='utf-8') as f:
    gpt_generate = json.load(f) 
```

```python
len(all_sentences)
```

```python
all_sentences_translation[0]
```

```python
gpt_generate[0]
```

```python
all_sentences = all_sentences + [g['chinese'] for g in gpt_generate] + [g['chinese'] for g in all_sentences_translation]
all_sentences = list(set(all_sentences))
print(len(all_sentences))
for g in gpt_generate[617:]:
    for i in range(len(all_sentences)):
        if g['chinese'] == all_sentences[i]:
            print(all_sentences.pop(i))
            break
all_sentences = list(set(all_sentences))
len(all_sentences)
```

```python
gpt_dataset = []
count = [0,0]
for i,s in enumerate(all_sentences):
    found = False
    emoji = ""
    for g in gpt_generate:
        if s == g['chinese']:
            found = True
            emoji = g['emoji']
            count[0] += 1
            break
    if found:
        gpt_dataset.append({'chinese': s, 'emoji': emoji, 'id': i})
        continue
    
    for g in all_sentences_translation:
        if s == g['chinese']:
            found = True
            count[1] += 1
            emoji = g['emoji']
            break
    if found:
        gpt_dataset.append({'chinese': s, 'emoji': emoji, 'id': i})

gpt_dataset[:10]    
```

```python
with open('../data/gpt_dataset.json', 'w', encoding='utf-8') as f:
    json.dump(gpt_dataset, f, ensure_ascii=False, indent=4)
```
