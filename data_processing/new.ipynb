```python
import json

with open('../data/all_sentences.json') as f:
    data = json.load(f)
```

```python
char_collection = {}
for s in data:
    for c in s:
        if c in char_collection:
            char_collection[c] += 1
        else:
            char_collection[c] = 1
```

```python
with open('../data/char_to_emoji.json') as f:
    char_to_emoji = json.load(f)
```

```python
with open('../data/zhuyin_to_char.json') as f:
    zhuyin_to_char = json.load(f)
```

```python
zhuyin_to_emoji = {}
for z in zhuyin_to_char:
    emojis = {}
    for c in zhuyin_to_char[z]:
        if c in char_to_emoji and char_to_emoji[c]['emoji']:
            e = char_to_emoji[c]['emoji'][0]
            score = 0
            if c in char_collection:
                score = char_collection[c]
            if e in emojis:
                emojis[e] += score
            else:
                emojis[e] = score
    zhuyin_to_emoji[z] = emojis
zhuyin_to_emoji
```

```python
def combine_emojis(a,b):
    for e in b:
        if e in a:
            a[e] += b[e]
        else:
            a[e] = b[e]
    return a
```

```python
with open('../data/chinese_chars.json') as f:
    chinese_chars = json.load(f)
```

```python
for cc in chinese_chars:
    c = cc['font']
    if c not in char_to_emoji:
        continue
    weighted_emojis = {}
    for z in cc['zhuyin']:
        if z in zhuyin_to_emoji:
            weighted_emojis = combine_emojis(weighted_emojis, zhuyin_to_emoji[z])
        else:
            print('missing zhuyin', z)
    char_to_emoji[c]['weighted_emojis'] = weighted_emojis
char_to_emoji
```

```python
with open('../data/char_to_emoji.json', 'w', encoding='utf-8') as f:
    json.dump(char_to_emoji, f, ensure_ascii=False, indent=4)
```
